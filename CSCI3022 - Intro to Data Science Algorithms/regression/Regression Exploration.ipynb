This is an exploration of different types of regression

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy
from sklearn import linear_model as linmod
import random
#Importing needed modules
```

This next cell is used to create the function to regress and some function to generate values with some noise

```python
def func1(x):
    #Returns a y value corresponding to the input x value
    return x*2

def noisyFunc1(x):
    return func1(x) + random.randint(-20,20)*random.random()

#Create some linspace as input for the 
x = numpy.linspace(-10,10,40)
```

Create a vector corresponding to the noisy data


```python
y = numpy.array([noisyFunc1(i) for i in x])
f = [func1(i) for i in x]
plt.figure()
plt.plot(x,y,'x')
plt.plot(x,func1(x),label='func1')
plt.legend(loc='lower left')
plt.show()
```

```python
print(x.shape)
print(y.shape)
```

```python
regr = linmod.LinearRegression()
regr.fit(x.reshape(40,-1),y.reshape(40,-1))

x_test = numpy.linspace(-10,10,20)
y_test = [noisyFunc1(i) for i in x_test]

#Copied from example on sklearn website:
# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % numpy.mean((regr.predict(x.reshape(40,-1)) - y.reshape(40,-1)) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x.reshape(40,-1), y.reshape(40,-1)))
```
